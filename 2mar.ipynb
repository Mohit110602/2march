```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1)
# Matplotlib is a library in Python used for visualizations pf the data. The plots in Matplotlib are: plot, scatter, bar, hist, pie, etc. 
```

```python
#2)
# A scatter plot is a type of plot having discrete coordinates as the graph.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Y vs X graph ')
plt.scatter(x,y)
```

```python
#3)
# Subplot() is used to plot multiple plots on a single graph.
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.title('Ist graph')
plt.plot(x1,y1)
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.title('2nd graph')
plt.plot(x2,y2)
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.title('3rd graph')
plt.plot(x3,y3)
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.title('4th graph')
plt.plot(x4,y4)
```

```python
#4)
# A Bar plot has the graph in the form of bar and is used to show the relation between numeric and categorical data.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel('Company')
plt.ylabel('Profit')
plt.bar(company,profit)
```

```python
#5)
# A box plot is a plot in which data is visualised in the form of boxes and whiskers. It is used to show the distribution of data.  
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(2,1,1)
plt.title('Ist plot')
plt.boxplot(box1)
plt.subplot(2,1,2)
plt.title('2nd plot')
plt.boxplot(box2)
```
